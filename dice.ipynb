```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Two dice rolling game
* In this game, there are two dices and our task is to calculate the sum of their face values. 
* In this notebook, we will observe the outcome of random values returned by the dices

```
face_values = pd.DataFrame(list(range(1,7)))
possible_outcome = face_values.sample(2,replace=True).sum().loc[0] #2 indicates two dices sample
print(possible_outcome,'is the sum of 2 dices')
```

```
print([face_values.sample(2,replace=True).sum().loc[0] for i in range(10)],'are the observation of rolling two dices 10 times')
```

Distribution for 100 outcomes

$Relative Frequency = Frequency/Number Of Trails$ 

```
outcome100 = pd.DataFrame([face_values.sample(2,replace=True).sum().loc[0] for i in range(100)])[0]
outcome100_relative_frequency = outcome100.value_counts().sort_index()/100
outcome100_relative_frequency.plot(kind='bar',figsize=(10,5),
                                   yticks = outcome100_relative_frequency.to_list())
plt.title('Distribution of 100 Observation')
plt.xlabel('Observations →')
_ = plt.ylabel('Relative Frequency →')
```

```
for index,value in enumerate([100,1000,10000]):
    plt.subplot(3,1,index+1,frameon=False)
    outcome = pd.DataFrame([face_values.sample(2,replace=True).sum().loc[0] for i in range(value)])[0]
    outcome_relative_frequency = outcome.value_counts().sort_index()/value
    outcome_relative_frequency.plot(kind='bar',figsize=(15,15),yticks = outcome100_relative_frequency.to_list())
    plt.title('Distribution of '+str(value)+' observation ↓')
    plt.ylabel('Relative Frequency →')
_ = plt.xlabel('Observations →')
```

**Quick Notes**
* Above distribution shows the frequency of distribution of outcomes
* As the number of outcomes increses, the distribution becomes more stable and approaching a normal distribution curve.
